```python
def finding_outlier(df):
  Q1 = df.quantile(0.25)
  Q3 = df.quantile(0.75)
  IQR  = Q3-Q1
  df_final = df[(df<(Q1-(1.5*IQR))) | (df>(Q3+(1.5*IQR)))]
  return df_final
```

```python
print(finding_outlier(df1['GDPperkapita']))
print(finding_outlier(df1['Pendapatan']))
```

```python
print(finding_outlier(df1['Kesehatan']))
print(finding_outlier(df1['Kematian_anak']))
```

```python
def remove_outlier(df):
  Q1 = df.quantile(0.25)
  Q3 = df.quantile(0.75)
  IQR  = Q3-Q1
  df_final = df[~(df>(Q1-(1.5*IQR))) | (df<(Q3+(1.5*IQR)))]
  return df_final
```

```python
df2 = remove_outlier(df1[['GDPperkapita', 'Pendapatan']])
df2.dropna(axis= 0, inplace=True)
df2
```

```python
df2_kesehatan = remove_outlier(df1[['Kesehatan', 'Kematian_anak']])
df2_kesehatan.dropna(axis= 0, inplace=True)
df2_kesehatan
```
